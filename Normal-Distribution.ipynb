#Normal (Gaussian) Distribution of heights



> X: height of a randomly picked person
>> X~Normal(mu, sigma)



```python
#Building a hypothetical population of heights, which are normally distributed

mu = 65
sigma = 2.5
heights = np.random.normal(loc=mu, scale=sigma, size=(3_000_000,))

print(f"The generated array of heights are = {np.round(heights,1)}")
```

```python
#Plotting the heights

plt.figure(figsize=(10,4))
plt.title(f"Probability Distribution of Heights (mu={mu}, sigma={sigma})")
plt.xlabel("Heights in inches -->")
plt.xticks(np.arange(50, 80+1, 2.5))
sns.histplot(x=heights,stat='density',color="lightskyblue")
plt.axvline(x=mu, color="black")


#+/- 1 sigma lines
plt.axvline(x=mu+1*sigma, color="blue")
plt.axvline(x=mu-1*sigma, color="blue")

#+/- 2 sigma lines
plt.axvline(x=mu+2*sigma, color="red")
plt.axvline(x=mu-2*sigma, color="red")

#+/- 3 sigma lines
plt.axvline(x=mu+3*sigma, color="green")
plt.axvline(x=mu-3*sigma, color="green")


plt.show()
```

```python
#What is the CDF of x = 65

from scipy import stats

x = 65

z = 0

stats.norm.cdf(z)   #input: z-score; output: cdf of z
```

```python
#What is the CDF of x = 62.5

x = 62.5

mu = 65
sigma = 2.5

z = (x - mu) / sigma

print(f"z-score = {z} and cdf =  {stats.norm.cdf(z)}")
```

```python
#What is the CDF of x = 67.5

x = 67.5

mu = 65
sigma = 2.5

z = (x - mu) / sigma

print(f"z-score = {z} and cdf =  {stats.norm.cdf(z)}")
```

#Tweaking the values of mu & sigma in

> X~Normal(mu,sigma)



```python
#Different values of (mu, sigma)

plt.figure(figsize=(18,12))

#subplot (2,2,1)
mu_1 = 65
sigma_1 = 2.5
heights_1 = np.random.normal(loc=mu_1, scale=sigma_1, size=(1_000_000,))

plt.subplot(2,2,1)
plt.title(f"X~Normal(mu={mu_1}, sigma={sigma_1})",fontsize=13)
plt.xlim(40,90)
plt.xticks(np.arange(50, 80+1, 5))
sns.histplot(x=heights_1,stat='density',color="lightskyblue")
plt.axvline(x=mu_1, color="black")

#subplot (2,2,2)
mu_2 = 50
sigma_2 = 2.5
heights_2 = np.random.normal(loc=mu_2, scale=sigma_2, size=(1_000_000,))

plt.subplot(2,2,2)
plt.title(f"X~Normal(mu={mu_2}, sigma={sigma_2})",fontsize=13)
plt.xlim(40,90)
plt.xticks(np.arange(50, 80+1, 5))
sns.histplot(x=heights_2,stat='density',color="lightskyblue")
plt.axvline(x=mu_1, color="black")
plt.axvline(x=mu_2, color="red")

#subplot (2,2,3)
mu_3 = 65
sigma_3 = 1
heights_3 = np.random.normal(loc=mu_3, scale=sigma_3, size=(1_000_000,))

plt.subplot(2,2,3)
plt.title(f"X~Normal(mu={mu_3}, sigma={sigma_3})",fontsize=13)
plt.xlim(40,90)
plt.xticks(np.arange(50, 80+1,5))
sns.histplot(x=heights_3,stat='density',color="lightskyblue")
plt.axvline(x=mu_3, color="black")

#subplot (2,2,4)
mu_4 = 65
sigma_4 = 15
heights_4 = np.random.normal(loc=mu_4, scale=sigma_4, size=(1_000_000,))

plt.subplot(2,2,4)
plt.title(f"X~Normal(mu={mu_4}, sigma={sigma_4})",fontsize=13)
plt.xlim(40,90)
plt.xticks(np.arange(50, 80+1, 5))
sns.histplot(x=heights_4,stat='density',color="lightskyblue")
plt.axvline(x=mu_4, color="black")


plt.show()
```
