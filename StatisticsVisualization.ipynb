```python
#importing libaries
#this is necessary for ALL code blocks below

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import binom
```

#Uniform Probability Distributions

##Tossing a fair coin

```python
#Showing the results graphically

N = 1000000 #number of trails / tosses

outcomes_in_N_tosses = np.random.choice(a = ["Heads","Tails"],
                                        size = N,
                                        p = [0.5,0.5]) #p = [prob of Tails, prob of Heads]

plt.title(f"Number of Tails and Heads in {N} tosses of a coin ")

sns.countplot(x=outcomes_in_N_tosses,
              order={"Heads","Tails"},
              color="crimson",
              stat="proportion") #stat = count / proportion / percent

plt.grid(axis = 'y')
plt.show()
```

##Rolling a fair dice

```python
#Rolling a die/dice N times and seeing the distribution of the Sample Space

N = 1000000 #number of rolls of a die

outcomes_in_N_rolls = np.random.choice(a = [1,2,3,4,5,6],
                                       size = N,
                                       p = [1/6,1/6,1/6,1/6,1/6,1/6]) #p = [probability distribution among outcomes]

plt.title(f"Distribution of die roll outcomes in {N} rolls")

sns.countplot(x = outcomes_in_N_rolls,
              order = {1,2,3,4,5,6},
              color = "green",
              stat = "proportion") #stat = count / proportion / percent

plt.grid(axis = 'y')
plt.show()
```

#Non-uniform Probability Distributions

```python
#X: Number of Heads in "n" tosses of a coin

#small "n" is the number of coin tosses in the experiment
n = 10 #number of trails / tosses in each round

#"R" is the number of times we run the experiment to view the distribution of outcomes
R = 4 #number of rounds of the experiment

outcomes_in_R_rounds = np.random.choice(a=[1,0], size=[R,n], p=[0.5,0.5])

print("Let 1 represent the outcome: 'Heads' \nand 0 represent the outcome: 'Tails'\n\n")

for r in range(len(outcomes_in_R_rounds)):
  print(f"Outcomes in round {r}: {outcomes_in_R_rounds[r]}")
  print(f"Number of heads in round {r} = {np.sum(outcomes_in_R_rounds[r])}\n\n")
```

##Binomial Distribution

```python
#X:Number of Heads in "n" tosses of a coin

#small "n" is the number of coin tosses in the experiment
n = 8 #number of trails / tosses in each round


#"R" is the number of times we run the experiment to view the distribution of outcomes
R = 1000000 #number of rounds of the experiment

outcomes_in_R_rounds = np.random.choice(a=[1,0], size=[R,n], p=[1/3,2/3]) #p = [prob of Heads, prob of Tails]

no_of_heads_in_each_round = np.sum(outcomes_in_R_rounds, axis = 1)

plt.figure(figsize=(10,5))
plt.title(f"Distribution of Number of Heads in {n} tosses; {R} rounds of experiment ")
plt.xlabel(f"No of Heads (k) in {n} tosses")
sns.countplot(x=no_of_heads_in_each_round, stat="proportion") #stat = count / proportion / percent
plt.grid(axis = 'y')
plt.show()
```

#Expected value of a binomial distribution

```python
#Expected value: weighted average
#weights are the probabilities for each outcome (value) in k

n = 10
#k = {0,1,2,...,n}
p = 0.5 #probability of success in a single trial


#initialize wt_avg to 0
wt_avg = 0

for i in range(n+1):
  wt_avg += i*binom.pmf(n=n, k=i, p=p)

print(f"Expected value = {np.round(wt_avg)}")
```

```python
#You subject 15 random (asymptomatic) people to a COVID
#Test accuracy is 90%
#What is the probability that 2 or more people will get a wrong test result?

#Showing the results graphically

N = 15 #number of bernoulli trials each round
R = 100000 #number of rounds of the experiment

#Let 0: "failure" and 1: "success" in each bernoulli trial
outcomes_in_R_rounds = np.random.choice(a=[0,1], size=[R,N], p=[0.9,0.1]) #p = [prob of failue, prob of success]

no_of_heads_in_each_round = np.sum(outcomes_in_R_rounds, axis = 1)

plt.figure(figsize=(10,5))
plt.title(f"Distribution of Number of Wrong Results in {N} tests; {R} rounds")
plt.xlabel(f"No of Wrong Results (K) in {N} tests")
sns.countplot(x=no_of_heads_in_each_round, stat="proportion") #stat = count / proportion / percent
plt.grid(axis = 'y')
plt.show()
```
