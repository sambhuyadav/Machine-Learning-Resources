```python
#importing libaries
#this is necessary for ALL code blocks below

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
```

#Penguin Data

```python
penguins = sns.load_dataset("penguins")

penguins = penguins.dropna()

penguins
```

## Weight vs Wingspan

```python
#Is there a relation b/w body_mass_g vs flipper_length_mm

plt.title("Weight vs Wingspan")

sns.scatterplot(data=penguins, x='body_mass_g', y='flipper_length_mm')

plt.show()
```

```python
#rho value

#Is there a relation b/w body_mass_g vs flipper_length_mm


#For Pearsonr test
#Ho: There is NO correlation between X and Y
#Ha: There is a correlation between X and Y

from scipy import stats

stats.pearsonr(x = penguins['body_mass_g'], y = penguins['flipper_length_mm'])



```

#Custom generated monotonic data

```python
#Positive monotonic function

x_1 = np.arange(1,5,0.5)
y_1 = x_1**2

x_2 = np.arange(5,15,0.5)
#y_2 = np.full(shape=x_2.shape, fill_value=25).astype('int')
y_2 = y_1[-1]+(x_2-5)/1000

x_3 = np.arange(15,20,0.5)
y_3 = y_2[-1]+(x_3-15)*4

x_vals = np.concatenate([x_1,x_2,x_3])
y_vals = np.concatenate([y_1,y_2,y_3])

plt.xlabel("x_vals")
plt.ylabel("y_vals")

pearson_rho = stats.pearsonr(x_vals, y_vals)[0]
spearman_rho = stats.spearmanr(x_vals, y_vals)[0]
plt.title(f"pearson_rho = {pearson_rho : .2f} and spearman_rho = {spearman_rho : .2f}")


plt.grid(axis="y")
sns.scatterplot(x=x_vals,y=y_vals)
plt.show()
```

```python
stats.pearsonr(x_vals, y_vals)[0]
```

```python
stats.spearmanr(x_vals, y_vals)[0]
```

```python
#Negative monotonic function


x_vals = np.arange(1,10,0.2)
w_vals = x_vals**5
y_vals = max(w_vals) - w_vals

plt.xlabel("x_vals")
plt.ylabel("y_vals")
pearson_rho = stats.pearsonr(x_vals, y_vals)[0]
spearman_rho = stats.spearmanr(x_vals, y_vals)[0]
plt.title(f"pearson_rho = {pearson_rho : .2f} and spearman_rho = {spearman_rho : .2f}")
plt.grid(axis="y")
sns.scatterplot(x=x_vals,y=y_vals)
plt.show()
```

#Heatmaps

```python
#Correlation Table

penguin_continous_RVs = penguins[['bill_length_mm', 'bill_depth_mm',
                           'flipper_length_mm', 'body_mass_g']]

print("Correlations between Categorical RVs of penguins dataset:\n\n\n")
penguin_continous_RVs.corr()
```

```python
#Heatmap of Correlation Table

plt.title("Heatmap of Correlations between Categorical RVs of penguins dataset\n")
plt.xticks(rotation=45)
sns.heatmap(data=penguin_continous_RVs.corr(), annot=True)
plt.show()
```
