#Chisquared: Goodness of Fit Test

```python
"""
I tossed a coin 10 times and it came up Heads 7 times.

Perform a Chisquared Goodness-of-fit test
to determine if it is a fair coin
"""

#Ho: observed frequencies fit with the expected
#Ha: observed frequencies do not fit with expected (are "significantly" different)

expected_frequencies = [5,5]

observed_frequencies = [7,3]

from scipy import stats

stats.chisquare(f_obs = observed_frequencies, f_exp = expected_frequencies) #observed freq vs expected frequencies


#p_val = 0.2 > alpha (0.05): Fail to Reject Ho
```

```python
"""
I tossed a coin 50 times and it came up Heads 28 times.

Perform a Chisquared Goodness-of-fit test
to determine if it is a fair coin
"""

#Ho: observed frequencies fit with the expected
#Ha: observed frequencies do not fit with expected (are "significantly" different)

expected_frequencies = [25,25]

observed_frequencies = [28,22]

from scipy import stats

stats.chisquare(f_obs = observed_frequencies, f_exp = expected_frequencies) #observed freq vs expected frequencies


#p_val = 0.2 > alpha (0.05): Fail to Reject Ho
```

```python
"""
I tossed a coin 50 times and it came up Heads 11 times.

Perform a Chisquared Goodness-of-fit test
to determine if it is a fair coin
"""

expected_frequencies = [25,25]

observed_frequencies = [11,39]


stats.chisquare(f_obs = observed_frequencies, f_exp = expected_frequencies) #observed freq vs expected frequencies


#p_val = 0.000075
```

```python
"""
In a city, people belong to 3 age groups: [children, adults, aged]
Here is the population split: [40%, 50%, 10% ]


Some new virus is spreading across the city, and you have been told
that 200 people are infected: out of which 60 are children and 35 are aged people



Perform a Chisquared Goodness-of-fit test
to determine if a person's age group has a significant their chance of catching the infection
"""

#Ho: the ratios will be similar to the ratios of the population
#Ha: the ratios of infected are DIFFERENT from the ratios of the population

expected_frequencies = [80, 100, 20]

observed_frequencies = [60, 105, 35]

stats.chisquare(f_obs = observed_frequencies, f_exp = expected_frequencies)
```
